### Matplotlib usage

#### Importing libraries

```python
from matplotlib import pyplot as plt
import numpy as np
```

#### Line graph to show position of two objects recorded at equal time intervals

```python
x1 = np.array([0, 6, 8, 4])
y1 = np.array([0, 2, 5, 8])
x2 = np.array([1, 3, 7, 9])
y2 = np.array([2, 6, 3, 7])

plt.plot(x1, y1, x2, y2, marker='x', linestyle='--')
plt.title("position of two objects")
plt.show()
```

#### bar graph of simple distribution

```python
x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])

plt.bar(x,y, width=0.4, color="green")
plt.title("no.of chocolates with each child")
plt.xlabel("name of child")
plt.show()
```

#### pie chart of top polluters of oceans

```python
y = np.array([29, 22, 14, 6])
mylabels = ["packaging", "food containers", "fishing and beach items", "glass bottles"]

plt.pie(y, labels = mylabels)
plt.title("top categories of polluters in oceans")
plt.show() 
```
