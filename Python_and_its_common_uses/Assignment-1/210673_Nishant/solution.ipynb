```python
#program to add two matrices without using numpy
import time
begin = time.time()
m1=int(input("enter rows:"))
n1=int(input("enter columns:"))
arr1=[[int(input()) for i in range (m1)]for j in  range (n1)]
m2=int(input("enter rows:"))
n2=int(input("enter columns:"))
if((m1!=m2) or (n1!=n2) ):
    print("Cannot Add")

else:
    arr2=[[int(input()) for i in range (m2)]for j in  range (n2)]
    arr3=[[0 for i in range (m1)]for j in  range (n2)]
    for i in range (m1):
        for j in range (n2):
            arr3[i][j]= arr1[i][j]+arr2[i][j]

    for i in range(m1):
        for j in range(n2):
            print(arr3[i][j], end = "")
        print()

end = time.time()
print(f"Total runtime of the program is {end - begin}")        
```

```python
#program to add two matrices using numpy
import numpy as np
import time
begin = time.time()

rows = int(input("Enter the Number of rows : " ))
column = int(input("Enter the Number of Columns: "))

print("Enter the elements of First Matrix:")
matrixa= [[int(input()) for i in range(column)] for i in range(rows)]
print("First Matrix is: ")
for n in matrixa:
    print(n)

matrixa = np.array(matrixa)

print("Enter the elements of Second Matrix:")
matrixb= [[int(input()) for i in range(column)] for i in range(rows)]
for n in matrixb:
    print(n)
    
matrixb = np.array(matrixb)    

result = np.add(matrixa, matrixb)

print("The Sum of Above two Matrices is : ")
for r in result:
    print(r)

end = time.time()
print(f"Total runtime of the program is {end - begin}") 

```

```python
#program to multiply two matrices without using numpy
import time
begin = time.time()
m1=int(input("enter rows:"))
n1=int(input("enter columns:"))
arr1=[[int(input()) for i in range (m1)]for j in  range (n1)]
m2=int(input("enter rows:"))
n2=int(input("enter columns:"))
if(n1!=m2):
    print("Can Multiply")

else:
    arr2=[[int(input()) for i in range (m2)]for j in  range (n2)]
    arr3=[[0 for i in range (m1)]for j in  range (n2)]
    for i in range (m1):
        for j in range (n2):
            for k in range (m1):
             arr3[i][j]=arr3[i][j]+ arr1[i][k]*arr2[k][j]

    for i in range(m1):
        for j in range(n2):
            print(arr3[i][j], end = "")
        print()

end = time.time()
print(f"Total runtime of the program is {end - begin}") 
```

```python
#program to multiply two matrices using numpy
import numpy as np
import time
begin = time.time()
rows = int(input("Enter the Number of rows : " ))
column = int(input("Enter the Number of Columns: "))

print("Enter the elements of First Matrix:")
matrixa= [[int(input()) for i in range(column)] for i in range(rows)]
print("First Matrix is: ")
for n in matrixa:
    print(n)

matrixa = np.array(matrixa)

print("Enter the elements of Second Matrix:")
matrixb= [[int(input()) for i in range(column)] for i in range(rows)]
for n in matrixb:
    print(n)
    
matrixb = np.array(matrixb)

res = np.dot(matrixa,matrixb)

print("The product of Above two Matrices is : ")
for r in res:
    print(r)

end = time.time()
print(f"Total runtime of the program is {end - begin}") 
```

```python
#program to inverse a matrix using numpy
import numpy as np
import time
begin = time.time()

rows = int(input("Enter the Number of rows : " ))
column = int(input("Enter the Number of Columns: "))

print("Enter the elements of Matrix:")
matrix= [[int(input()) for i in range(column)] for i in range(rows)]
print("Matrix is: ")
for n in matrix:
    print(n)

matrix = np.array(matrix)

print(np.linalg.inv(matrix))

end = time.time()
print(f"Total runtime of the program is {end - begin}") 
```
