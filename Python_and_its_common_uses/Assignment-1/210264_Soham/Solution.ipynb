## Matrix multiplication

```python
import numpy as np
import time
# taking size of matrices
a=np.random.randint(2,10)
b=np.random.randint(2,10)
c=np.random.randint(2,10)
#making random matrices
arr1=np.random.randn(a,b)
arr2=np.random.randn(b,c)
arr3=np.zeros([a,c])
#marking initial time
time1=time.time()
#multiplying the arrays 1 and 2 and storing the result in array 3
for x in range(len(arr1)):
    for y in range(len(arr2[0])):
        for z in range(len(arr2)):
            arr3[x,y]+=arr1[x,z]*arr2[z,y]
print(arr3)
time2=time.time() #end of my program time = start of the library program
print(arr1.dot(arr2)) #multiplication
time3=time.time() #end of library program
print(time2-time1)
print(time3-time2)
```

__we can see that the answer of matrix multiplication is the same but the time taken while using for loops(in the program made by me) is different__
## Matrix addition

```python
#few more random arrays for addition
arr6=np.zeros([a,b])
arr4=np.random.randn(a,b)
arr5=np.random.randn(a,b)
time1=time.time() # start time
#matrix summation
for x in range(len(arr4)):
    for y in range(len(arr4[0])):
        arr6[x,y]=arr4[x,y]+arr5[x,y]
print(arr6)
time2=time.time() #end time 1 
print(arr4+arr5)
time3=time.time() #end time 2
print(time2-time1)
print(time3-time2)
```

## Inverse of a matrix

```python
time1=time.time()
def smatrix(arr, x, y, n): #returns the matrix we get on removing the row and column x,y resp.
    ar=np.zeros([n-1,n-1])
    for i in range(n):
        for j in range(n):
            if (i==x or j==y):
                continue
            elif j<y and i<x:
                ar[i][j]=arr[i][j]
            elif j>y and i>x:
                ar[i-1][j-1]=arr[i][j]
            elif j>y and i<x:
                ar[i][j-1]=arr[i][j]
            elif j<y and i>x:
                ar[i-1][j]=arr[i][j]
    return ar

def determinant(arr): #for getting determinant of the matrix using recursion
    if len(arr)==2:
        return arr[0][0]*arr[1][1]-arr[0][1]*arr[1][0]
    det=0
    for x in range(len(arr)):
        det+=((-1)**x)*arr[0][x]*determinant(smatrix(arr,0, x,len(arr)))
    return det

arr=np.random.randn(a,a)
arans=np.zeros([a,a])
detarr=determinant(arr)
if detarr==0:
    print("determinant is zero hence inverse not possible")
#making the program for finding cofactors and storing them in appropriate places for adjecent matrix and dividing the value by determinant to get ht eentries in inverse of the matrix
for x in range(a):
    for y in range(a):
        arans[y][x]=((-1)^(x+y))*determinant(smatrix(arr, x, y, a))/detarr
        arans[y][x]/=x+y+1
        if (x+y)%2 == 0:
            arans[y][x]*=-1
detarans=determinant(arans)
print(detarans)
print(detarr)
print(arr)
print(arans)
time2=time.time()
print(np.linalg.inv(arr)) #returns inverse of the inputted matrix
time3=time.time()
print(time2-time1)
print(time3-time2)
```
